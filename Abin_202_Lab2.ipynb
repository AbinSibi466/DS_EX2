```python
import numpy as np
import matplotlib.pyplot as plt

# Step function definition
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Generate input data
x_values = np.linspace(-10, 10, 400)

# Apply step function
y_values = step_function(x_values)

# Plot the step function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="Step Function", color='blue')
plt.title("Step Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.axhline(0, color="black",linewidth=0.5)
plt.axvline(0, color="black",linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Binary Sigmoid (Logistic Sigmoid) function
def binary_sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Bipolar Sigmoid (Tanh) function
def bipolar_sigmoid(x):
    return np.tanh(x)

# Generate input data
x_values = np.linspace(-10, 10, 400)

# Apply both sigmoid functions
binary_sigmoid_values = binary_sigmoid(x_values)
bipolar_sigmoid_values = bipolar_sigmoid(x_values)

# Plot both sigmoid functions
plt.figure(figsize=(10, 6))

# Plot Binary Sigmoid
plt.subplot(1, 2, 1)
plt.plot(x_values, binary_sigmoid_values, label="Binary Sigmoid", color='blue')
plt.title("Binary Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.axhline(0, color="black",linewidth=0.5)
plt.axvline(0, color="black",linewidth=0.5)
plt.legend()

# Plot Bipolar Sigmoid (Tanh)
plt.subplot(1, 2, 2)
plt.plot(x_values, bipolar_sigmoid_values, label="Bipolar Sigmoid (Tanh)", color='green')
plt.title("Bipolar Sigmoid (Tanh) Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.axhline(0, color="black",linewidth=0.5)
plt.axvline(0, color="black",linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh function implementation
def tanh_function(x):
    return np.tanh(x)

# Generate input data
x_values = np.linspace(-10, 10, 400)

# Apply tanh function
tanh_values = tanh_function(x_values)

# Plot the Tanh function
plt.figure(figsize=(8, 6))
plt.plot(x_values, tanh_values, label="Tanh Function", color='purple')
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU function implementation
def relu_function(x):
    return np.maximum(0, x)

# Generate input data
x_values = np.linspace(-10, 10, 400)

# Apply ReLU function
relu_values = relu_function(x_values)

# Plot the ReLU function
plt.figure(figsize=(8, 6))
plt.plot(x_values, relu_values, label="ReLU Function", color='orange')
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Sample data: XOR problem (for binary classification)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python
# Function to create a neural network model with different activation functions
def create_nn(activation_func):
    model = Sequential()
    # Input layer with 2 features and hidden layer with 8 neurons
    model.add(Dense(8, input_dim=2, activation=activation_func))
    # Output layer with 1 neuron (binary classification) using sigmoid
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Create model with Sigmoid activation
model_sigmoid = create_nn('sigmoid')

# Train the model
print("\nTraining model with Sigmoid activation:")
model_sigmoid.fit(X, y, epochs=100, verbose=0)

# Evaluate the model
loss_sigmoid, accuracy_sigmoid = model_sigmoid.evaluate(X, y, verbose=0)
print(f"Sigmoid Model Accuracy: {accuracy_sigmoid*100:.2f}%")
```

```python
# Create model with Tanh activation
model_tanh = create_nn('tanh')

# Train the model
print("\nTraining model with Tanh activation:")
model_tanh.fit(X, y, epochs=100, verbose=0)

# Evaluate the model
loss_tanh, accuracy_tanh = model_tanh.evaluate(X, y, verbose=0)
print(f"Tanh Model Accuracy: {accuracy_tanh*100:.2f}%")
```

```python
# Create model with ReLU activation
model_relu = create_nn('relu')

# Train the model
print("\nTraining model with ReLU activation:")
model_relu.fit(X, y, epochs=100, verbose=0)

# Evaluate the model
loss_relu, accuracy_relu = model_relu.evaluate(X, y, verbose=0)
print(f"ReLU Model Accuracy: {accuracy_relu*100:.2f}%")
```
